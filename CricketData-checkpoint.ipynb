```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("matches.csv")
```

```python
# print(df[df.season == 2010].city)
# df.describe()
# df.info()
df.head()
```

```python
df.isnull().sum()
```

DROP umpire3 COLUMN

```python
df = df.drop("umpire3", axis=1)
df.head()
```

```python
plt.figure(figsize=(6, 6))
df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Toss Decision Distribution')
plt.ylabel('')
plt.show()
```

```python
# Count the number of matches won and lost for each toss decision
toss_decision_counts = df['toss_decision'].value_counts()
toss_decision_win_counts = df[df['result'] == 'normal'].groupby('toss_decision')['result'].count()
toss_decision_lose_counts = toss_decision_counts - toss_decision_win_counts

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=toss_decision_counts.index, y=toss_decision_win_counts, label='Win', color='lightblue')
sns.barplot(x=toss_decision_counts.index, y=toss_decision_lose_counts, label='Lose', bottom=toss_decision_win_counts, color='salmon')
plt.title('Matches Won and Lost by Toss Decision')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Matches')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='winner', data=df, palette='Set2', order=df['winner'].value_counts().index)
plt.title('Number of Wins by Each Team')
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='win_by_runs', data=df, color='lightblue')
plt.title('Distribution of Wins by Runs ')
plt.xlabel('Win by Runs')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['win_by_wickets'], bins=10, kde=True, color='salmon')
plt.title('Distribution of Wins by Wickets)')
plt.xlabel('Win by Wickets')
plt.ylabel('Frequency')
plt.show()
```

```python
top_players = df['player_of_match'].value_counts().head(6)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_players.index, y=top_players.values, palette='Set3')
plt.title('Highest Man of the Match Award Winners')
plt.xticks(rotation=90)
plt.xlabel('Players')
plt.ylabel('Number of Awards')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='city', data=df, palette='viridis')
plt.title('Number of Matches Played in Each City')
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.show()
```
